```python
#Trying many models to figure out which one gives the best score using GridSearchCV
from sklearn.model_selection import ShuffleSplit
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import GridSearchCV
from sklearn.linear_model import Lasso
from sklearn.linear_model import Ridge
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.ensemble import VotingRegressor
from sklearn.ensemble import StackingRegressor
from catboost import CatBoostRegressor
from lightgbm import LGBMRegressor
from xgboost import XGBRegressor
```

```python
#We now write a function

def find_best_model_using_gridsearchcv(X,y):
    algos = {
        "linear_regression":{
            "model":LinearRegression(),
            "params":{
                "normalize":[True, False]
            }
        },
        
        "ridge": {
            "model": Ridge(),
            "params": {
                "alpha": [1,2],
                "normalize":[True, False]
            }
        },
                
        "lasso": {
            "model": Lasso(),
            "params": {
                "alpha": [1,2],
                "selection": ["random", "cyclic"]
            }
        },
        
        "decision_tree": {
            "model": DecisionTreeRegressor(),
            "params": {
                "criterion": ["mse", "friedman_mse"],
                "splitter": ["best", "random"]
            }
        },
        
        "catboost":{
            "model": CatBoostRegressor(),
            "params":{
                'depth' :[6,8,10],
                'learning_rate' :[0.01,0.05,0.1],
                'iterations' :[30,50,100]
            }
        },
        
        "XGBreg":{
            "model": XGBRegressor(),
            "params":{
                'random_state':[42, 45,100]
            }
        },
        
        "randomforest":{
            "model":RandomForestRegressor(),
            "params":{
                'random_state':[42,45,100]
            }
        }
    }
    
    #To save the model into a dataframe
    
    scores = []
    cv = ShuffleSplit(n_splits=5, test_size=0.2, random_state=42)
    for algo_name, config in algos.items():
        gs = GridSearchCV(config["model"], config["params"], cv=cv, return_train_score=False)
        gs.fit(X,y)
        scores.append({
            "model":algo_name,
            "best_score":gs.best_score_,
            "best_params":gs.best_params_
        })
        
    return pd.DataFrame(scores, columns=["model","best_score", "best_params"])

# Calling the function

find_best_model_using_gridsearchcv(X,y)
    
```

```python
# stacking the best three model
r1 = XGBRegressor(random_state = 42)
r2 = RandomForestRegressor(random_state=100)
r3 = CatBoostRegressor(verbose=False,random_state = 42)
```

```python
model = VotingRegressor([('r1', r1),('r2', r2),('r3',r3)],weights=(1,1,1))
model.fit(X,y)
val_pred = model.predict(X_test)
rmse = mean_squared_error(y_test, val_pred, squared=False)
print("RMSE:", rmse)
```

```python
print(f'Accuracy score: {model.score(X_test, y_test)}')
```
